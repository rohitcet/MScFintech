```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch NASDAQ stock symbols using yfinance
# yfinance does not provide direct `get_nasdaq_symbols`. You can fetch info as shown below
all_ticker = yf.Ticker("AAPL")  # Example for a single ticker

# Print basic information about the ticker
print(all_ticker.info)  # This prints the metadata for AAPL stock

# Fetch historical data for AAPL
aapl = yf.download("AAPL", start="2015-01-01", end="2015-12-31")

# Print the dataframe info
print(aapl.info())
print(aapl)

# Plot the closing prices
plt.figure(figsize=(10, 6))
plt.plot(aapl['Close'], label="AAPL Closing Prices")
plt.title("AAPL Closing Prices in 2015")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```
