```python
# Step 1: Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Step 2: Load the dataset
# Assuming the 'train.csv' file is in the same directory as the script
data = pd.read_csv(r"C:\Users\S4 User\Desktop\Required Assignment 17.1_Classification with scikit-learn   [Rohit K Sharma]\train.csv")

# Step 3: Plot binary distribution of 'CentralAir' using seaborn catplot
#sns.catplot(x='CentralAir', data=data, kind='count', palette='pastel')
sns.catplot(x='CentralAir', data=data, kind='count', hue='CentralAir', palette='pastel')
plt.title("Distribution of 'CentralAir' (Y vs N)")
plt.show()

# Step 4: Plot 'CentralAir' vs 'SalePrice' using seaborn boxplot
#sns.boxplot(x='CentralAir', y='SalePrice', data=data, palette='pastel')
sns.boxplot(x='CentralAir', y='SalePrice', data=data, hue='CentralAir', palette='pastel')
plt.title("CentralAir vs SalePrice")
plt.show()

# Step 5: Partition the dataset into training and test sets
# For this task, we create a train-test split based on the given row count (1100 for training, 360 for testing)
houseData_train = data.iloc[:1100]
houseData_test = data.iloc[1100:1460]  # Remaining rows for test dataset

# Step 6: Convert 'CentralAir' to binary (Y:1, N:0)
#houseData_train['CentralAir'] = houseData_train['CentralAir'].map({'Y': 1, 'N': 0})
#houseData_test['CentralAir'] = houseData_test['CentralAir'].map({'Y': 1, 'N': 0})

# Creating explicit copies of the DataFrames
houseData_train = houseData_train.copy()
houseData_test = houseData_test.copy()

# Mapping values of 'CentralAir' from 'Y' and 'N' to 1 and 0
houseData_train['CentralAir'] = houseData_train['CentralAir'].map({'Y': 1, 'N': 0})
houseData_test['CentralAir'] = houseData_test['CentralAir'].map({'Y': 1, 'N': 0})


# Step 7: Training: Fit Decision Tree Model for 'CentralAir' prediction using 'SalePrice'
X_train = houseData_train[['SalePrice']]  # Features (SalePrice)
y_train = houseData_train['CentralAir']   # Target (CentralAir)

# Fit the model
model_saleprice = DecisionTreeClassifier(random_state=42)
model_saleprice.fit(X_train, y_train)

# Step 8: Predict 'CentralAir' for the test dataset and print out confusion matrix
X_test = houseData_test[['SalePrice']]
y_test = houseData_test['CentralAir']

# Predictions
y_pred_saleprice = model_saleprice.predict(X_test)

# Confusion matrix and classification metrics for SalePrice model
conf_matrix_saleprice = confusion_matrix(y_test, y_pred_saleprice)
class_report_saleprice = classification_report(y_test, y_pred_saleprice)

# Print Confusion Matrix and Classification Report
print("Confusion Matrix for 'SalePrice':")
print(conf_matrix_saleprice)
print("\nClassification Report for 'SalePrice':")
print(class_report_saleprice)

# Step 9: Repeat the same for other features - 'GrLivArea'
X_train_grliv = houseData_train[['GrLivArea']]
y_train_grliv = houseData_train['CentralAir']

# Train model on 'GrLivArea'
model_grliv = DecisionTreeClassifier(random_state=42)
model_grliv.fit(X_train_grliv, y_train_grliv)

# Predict 'CentralAir' for test data based on 'GrLivArea'
X_test_grliv = houseData_test[['GrLivArea']]
y_pred_grliv = model_grliv.predict(X_test_grliv)

# Confusion matrix and classification report for GrLivArea model
conf_matrix_grliv = confusion_matrix(y_test, y_pred_grliv)
class_report_grliv = classification_report(y_test, y_pred_grliv)

# Print Confusion Matrix and Classification Report
print("Confusion Matrix for 'GrLivArea':")
print(conf_matrix_grliv)
print("\nClassification Report for 'GrLivArea':")
print(class_report_grliv)

# Step 10: Repeat for 'LotArea'
X_train_lot = houseData_train[['LotArea']]
y_train_lot = houseData_train['CentralAir']

# Train model on 'LotArea'
model_lot = DecisionTreeClassifier(random_state=42)
model_lot.fit(X_train_lot, y_train_lot)

# Predict 'CentralAir' for test data based on 'LotArea'
X_test_lot = houseData_test[['LotArea']]
y_pred_lot = model_lot.predict(X_test_lot)

# Confusion matrix and classification report for LotArea model
conf_matrix_lot = confusion_matrix(y_test, y_pred_lot)
class_report_lot = classification_report(y_test, y_pred_lot)

# Print Confusion Matrix and Classification Report
print("Confusion Matrix for 'LotArea':")
print(conf_matrix_lot)
print("\nClassification Report for 'LotArea':")
print(class_report_lot)

# Step 11: Repeat for 'TotalBsmtSF'
X_train_bsmt = houseData_train[['TotalBsmtSF']]
y_train_bsmt = houseData_train['CentralAir']

# Train model on 'TotalBsmtSF'
model_bsmt = DecisionTreeClassifier(random_state=42)
model_bsmt.fit(X_train_bsmt, y_train_bsmt)

# Predict 'CentralAir' for test data based on 'TotalBsmtSF'
X_test_bsmt = houseData_test[['TotalBsmtSF']]
y_pred_bsmt = model_bsmt.predict(X_test_bsmt)

# Confusion matrix and classification report for TotalBsmtSF model
conf_matrix_bsmt = confusion_matrix(y_test, y_pred_bsmt)
class_report_bsmt = classification_report(y_test, y_pred_bsmt)

# Print Confusion Matrix and Classification Report
print("Confusion Matrix for 'TotalBsmtSF':")
print(conf_matrix_bsmt)
print("\nClassification Report for 'TotalBsmtSF':")
print(class_report_bsmt)

# Step 12: Combine all features ('SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF') for the Decision Tree model
X_train_all = houseData_train[['SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF']]
y_train_all = houseData_train['CentralAir']

# Train model on all features
model_all = DecisionTreeClassifier(random_state=42)
model_all.fit(X_train_all, y_train_all)

# Predict 'CentralAir' for test data based on all features
X_test_all = houseData_test[['SalePrice', 'GrLivArea', 'LotArea', 'TotalBsmtSF']]
y_pred_all = model_all.predict(X_test_all)

# Confusion matrix and classification report for model with all features
conf_matrix_all = confusion_matrix(y_test, y_pred_all)
class_report_all = classification_report(y_test, y_pred_all)

# Print Confusion Matrix and Classification Report
print("Confusion Matrix for 'All Features':")
print(conf_matrix_all)
print("\nClassification Report for 'All Features':")
print(class_report_all)

# Step 13: Comparison and Discussion
# Calculate and compare accuracy, true positives, false positives, etc.
accuracy_saleprice = conf_matrix_saleprice.diagonal().sum() / conf_matrix_saleprice.sum()
accuracy_grliv = conf_matrix_grliv.diagonal().sum() / conf_matrix_grliv.sum()
accuracy_lot = conf_matrix_lot.diagonal().sum() / conf_matrix_lot.sum()
accuracy_bsmt = conf_matrix_bsmt.diagonal().sum() / conf_matrix_bsmt.sum()
accuracy_all = conf_matrix_all.diagonal().sum() / conf_matrix_all.sum()

print("\nAccuracy Comparison:")
print(f"Accuracy for 'SalePrice' model: {accuracy_saleprice:.2f}")
print(f"Accuracy for 'GrLivArea' model: {accuracy_grliv:.2f}")
print(f"Accuracy for 'LotArea' model: {accuracy_lot:.2f}")
print(f"Accuracy for 'TotalBsmtSF' model: {accuracy_bsmt:.2f}")
print(f"Accuracy for 'All Features' model: {accuracy_all:.2f}")

# Step 14: False Positive and False Negative Discussion
print("\nFalse Positive/Negative Rates Analysis:")
for feature, cm in [('SalePrice', conf_matrix_saleprice), 
                    ('GrLivArea', conf_matrix_grliv),
                    ('LotArea', conf_matrix_lot), 
                    ('TotalBsmtSF', conf_matrix_bsmt), 
                    ('All Features', conf_matrix_all)]:
    fp_rate = cm[0, 1] / (cm[0, 1] + cm[1, 1]) if (cm[0, 1] + cm[1, 1]) != 0 else 0
    fn_rate = cm[1, 0] / (cm[1, 0] + cm[0, 0]) if (cm[1, 0] + cm[0, 0]) != 0 else 0
    print(f"\nFor {feature} model:")
    print(f"False Positive Rate: {fp_rate:.2f}")
    print(f"False Negative Rate: {fn_rate:.2f}")

# Discuss if False Positives and False Negatives are impacted by unbalanced classes
print("\nDiscussion on False Positives and False Negatives:")
# Here you can add any specific insights related to class imbalance and how it affects the performance of the models.
```
