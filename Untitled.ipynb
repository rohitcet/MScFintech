```python
import yfinance as yf
import mplfinance as mpf
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd

# Function to take user input for selecting options
def user_input():
    word = '''
Welcome to Stock Master!

Please select a function to continue:

(1) Search for a stock code by keyword
(2) View stock trend by code

(sample input 1 or 2)
'''

    next_step = '1'
    while next_step == '1':
        function = str(input(word))

        if function == '1':
            check_code()
        else:
            view_trend()

        next_step = next_request()
    else:
        print("Thank you for using the system. Bye-bye.")

# Function to check stock code based on a keyword
def check_code():
    # Ask user to input keyword for stock search
    input_keyword = input('Please input keyword for stock searching for: ')

    # Fetch available symbols manually or from a source (example with a small list here)
    # Normally you'd fetch this from an API or CSV, but for simplicity, we're hardcoding some common stock symbols
    all_ticker = {
        'AAPL': 'Apple Inc.',
        'GOOGL': 'Alphabet Inc.',
        'MSFT': 'Microsoft Corporation',
        'AMZN': 'Amazon.com Inc.',
        'TSLA': 'Tesla Inc.',
    }

    # Search for matching symbols based on the user's input
    selected_ticker = {symbol: name for symbol, name in all_ticker.items() if input_keyword.lower() in name.lower()}
    
    # Display the selected stock symbols
    if selected_ticker:
        print("Matching stocks found:")
        for symbol, name in selected_ticker.items():
            print(f"{symbol}: {name}")
    else:
        print("No matching stocks found.")

# Function to view stock trend based on user input
def view_trend():
    # Ask user to input stock code
    code = str(input('Please input the code of the stock requested:\n'))

    # Ask user to input start and end dates, and validate them
    while True:
        try:
            input_start = str(input('Please enter start date (eg. 2019-1-1): '))
            input_end = str(input('Please enter end date (eg. 2019-1-1): '))
            
            # Convert string input to datetime
            start = datetime.strptime(input_start, "%Y-%m-%d")
            end = datetime.strptime(input_end, "%Y-%m-%d")
            
            # Get source data of historical stock prices using yfinance
            df = yf.download(code, start=start, end=end)
            break
        except:
            print('The format of the date you entered is incorrect. Please try again.')

    # Ensure necessary columns are present and correctly formatted
    required_columns = ['Open', 'High', 'Low', 'Close', 'Volume']
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        print(f"Missing columns: {missing_columns}")
        return  # Exit if there are missing columns

    # Flatten the dataframe columns: remove the multi-level column index if present
    df.columns = df.columns.get_level_values(0)  # Get the top level of the multi-column index
    
    # Convert relevant columns to numeric values to prevent errors in plotting
    df[['Open', 'High', 'Low', 'Close', 'Volume']] = df[['Open', 'High', 'Low', 'Close', 'Volume']].apply(pd.to_numeric, errors='coerce')

    # Check data types of the columns
    print("\nData types after conversion:")
    print(df[['Open', 'High', 'Low', 'Close', 'Volume']].dtypes)

    # Check if 'df' is correctly loaded with columns
    print("\nFirst 4 rows of the stock data:")
    print(df.head(4))  # Display the first 4 rows to ensure data format is correct

    # Ask for graph choices
    graph_choice = str(input('''
Please select type of graph.
(1) Candle Stick (default: 'ohlc')
(2) Moving Average
(3) Volume

You may select multiple choices, separate each choice by comma.

Sample input for candle stick + Moving Average + Volume: 1,2,3
'''))

    # Manipulate choice input (convert to list of choices)
    graph_choice = graph_choice.split(',')

    # Draw graph based on user's input
    for choice in graph_choice:
        if choice == '1':
            try:
                # Reformat the data for mplfinance
                df_mpf = df[['Open', 'High', 'Low', 'Close', 'Volume']]
                df_mpf.index = pd.to_datetime(df_mpf.index)  # Ensure datetime index

                # Plot candlestick using 'mplfinance'
                mpf.plot(df_mpf, type='candle', title=code, style='charles', ylabel='Price ($)')
            except ValueError as e:
                print(f"Error plotting candlestick: {e}")
        elif choice == '2':
            # Calculate moving average (default 30 days)
            moving_average = df['Close'].rolling(window=30).mean()
            plt.plot(moving_average, label='30-day Moving Average')
        elif choice == '3':
            # Plot volume
            plt.bar(df.index, df['Volume'], label='Volume', width=0.5)

    # Finalize and display the plot for moving average and volume
    if '2' in graph_choice or '3' in graph_choice:
        plt.title(f"Stock Trend for {code}")
        plt.xlabel("Date")
        plt.ylabel("Values")
        plt.legend()
        plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
        plt.tight_layout()  # Ensure layout doesn't overlap
        plt.show()

# Function to ask the user for the next step
def next_request():
    next_word = '''
Would you like to
1. Return to menu
2. Exit program
(sample input : 1 or 2)
'''

    next_step = str(input(next_word))
    return next_step

# Main function to start the program
def main():
    user_input()

if __name__ == "__main__":
    main()
```
