```python
import pandas as pd

# Read the CSV file from the specified path
print("Reading the dataset from the provided path...")
data = pd.read_csv(r'C:\Users\S4 User\Advertising.csv', index_col=0)
print("Dataset loaded successfully!")

# Display the first 5 rows
print("\nPreview of the first 5 rows of the dataset:")
print(data.head())

# Display the last 5 rows
print("\nPreview of the last 5 rows of the dataset:")
print(data.tail())

# Check the shape of the DataFrame (number of rows and columns)
print("\nShape of the dataset (number of rows, columns):")
print(data.shape)

# Import necessary libraries for visualization
print("\nImporting Matplotlib and Seaborn for data visualization...")
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the relationship between features and response
print("\nCreating scatterplots to visualize relationships between features (TV, Radio, Newspaper) and Sales...")
sns.pairplot(data, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=7, aspect=0.7, kind='reg')
# plt.show()

# Create a Python list of feature names
print("\nSelecting feature columns for regression analysis...")
feature_cols = ['TV', 'Radio', 'Newspaper']
print("Selected features:", feature_cols)

# Use the list to select a subset of the original DataFrame (feature matrix)
print("\nExtracting the feature matrix (X) using selected columns...")
X = data[feature_cols]
print("\nPreview of the feature matrix (X):")
print(X.head())

# Extract the response vector (y) from the DataFrame
print("\nExtracting the response vector (y) containing the 'Sales' column...")
y = data['Sales']
print("\nPreview of the response vector (y):")
print(y.head())

# Split the dataset into training and testing sets
print("\nSplitting the dataset into training and testing sets...")
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print("Dataset successfully split into training and testing sets!")
print("Shape of training set (X_train, y_train):", X_train.shape, y_train.shape)
print("Shape of testing set (X_test, y_test):", X_test.shape, y_test.shape)

# Import Linear Regression model
print("\nImporting Linear Regression model...")
from sklearn.linear_model import LinearRegression

# Instantiate the model
print("Instantiating the Linear Regression model...")
model = LinearRegression()

# Fit the model to the training data
print("\nFitting the model to the training data...")
model.fit(X_train, y_train)
print("Model training completed!")

# Display feature names with their corresponding coefficients
print("\nFeature names and their corresponding coefficients:")
print(list(zip(feature_cols, model.coef_)))

# Make predictions on the testing set
print("\nMaking predictions on the testing set...")
y_pred = model.predict(X_test)

# Calculate Mean Absolute Error (MAE)
print("\nCalculating Mean Absolute Error (MAE)...")
from sklearn import metrics
true = [100, 50, 30, 20]
pred = [90, 50, 50, 30]
mae_hand = (10 + 0 + 20 + 10) / 4.0
print("MAE calculated by hand:", mae_hand)
print("MAE calculated using scikit-learn:", metrics.mean_absolute_error(true, pred))

# Calculate Mean Squared Error (MSE)
print("\nCalculating Mean Squared Error (MSE)...")
mse_hand = (10**2 + 0**2 + 20**2 + 10**2) / 4.0
print("MSE calculated by hand:", mse_hand)
print("MSE calculated using scikit-learn:", metrics.mean_squared_error(true, pred))

# Calculate Root Mean Squared Error (RMSE)
print("\nCalculating Root Mean Squared Error (RMSE)...")
import numpy as np
rmse_hand = np.sqrt(mse_hand)
print("RMSE calculated by hand:", rmse_hand)
print("RMSE calculated using scikit-learn:", np.sqrt(metrics.mean_squared_error(true, pred)))

# Calculate RMSE on actual predictions
print("\nCalculating RMSE on the test set predictions...")
rmse_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print("RMSE on test predictions:", rmse_test)

# Re-run regression with only two features (TV, Radio)
print("\nRepeating regression with only two features: TV and Radio...")
feature_cols = ['TV', 'Radio']
X = data[feature_cols]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

print("\nFitting the model with the reduced feature set...")
model.fit(X_train, y_train)

print("\nMaking predictions on the test set with the reduced feature set...")
y_pred = model.predict(X_test)

print("\nComputing RMSE for the reduced feature set...")
rmse_reduced = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print("RMSE with reduced feature set:", rmse_reduced)

```

```python
import pandas as pds
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import matthews_corrcoef

# Load the dataset
pima = pds.read_table(r"C:\Users\S4 User\pima.txt", sep="\t", header=0)
print(pima.shape)
print(pima.columns)
print(pima.dtypes)
print(pima.head(4))

# Separate features and target
data = pima.values
X = data[:, 0:8]  # Features
y = data[:, 8]    # Target

# Split the data into train and test sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, y, test_size=200, random_state=0
)

# Scale the features for better model performance
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the Logistic Regression model
lr = LogisticRegression(max_iter=500, solver='lbfgs')  # Increase max_iter
model = lr.fit(X_train_scaled, y_train)

# Display the coefficients and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Make predictions on the test data
y_predict = model.predict(X_test_scaled)
print("Predictions:", y_predict)

# Calculate and display the accuracy score
score = model.score(X_test_scaled, y_test)
print("Accuracy Score:", score)

# Calculate and display the MCC value
mcc = matthews_corrcoef(y_test, y_predict)
print("Matthews Correlation Coefficient:", mcc)

# Predict probabilities
y_proba = model.predict_proba(X_test_scaled)
print("Predicted Probabilities:", y_proba)

# calculate the MCC value
from sklearn.metrics import matthews_corrcoef
mcc = matthews_corrcoef(y_test, y_predict)
print(mcc)

# print the first 10 predicted responses
print(model.predict(X_test)[0:10])

# print the first 10 predicted probabilities of class membership
print(model.predict_proba(X_test)[0:10, :])

# print the first 10 predicted probabilities for class 1
print(model.predict_proba(X_test)[0:10, 1])

# store the predicted probabilities for class 1
y_pred_prob = model.predict_proba(X_test)[:, 1]

import matplotlib.pyplot as plt
from sklearn import metrics

# IMPORTANT: first argument is true values, second argument is predicted probabilities
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for diabetes classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
plt.show()

```
